```python
import pandas as pd
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
#https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html     -> Documentation for pyplot
```

```python
df = pd.read_excel('Data.xlsx')
df
```

```python
year =df['Year']
print(year)
```

```python
sales = df['Revenue']
print(sales)
```

```python
plt.figure(figsize = (15, 10))
plt.plot(year, sales, color = 'g')
plt.title('Sales Per Year')
plt.xlabel('Year')
plt.ylabel('Sales ($)')
plt.legend(['Sales'])
plt.show()
```

```python
#Challenge
test = pd.read_csv('names.csv')
test.head()
```

```python
name = 'Anna'                        #Assigning desired values to variables is the most efficient method
gender = 'F'
state = 'CA'
```

```python
a = test[test['Name'] == name]              #Obtaining the required information using variables created above
b = a[a['Gender'] == gender]
c = b[b['State'] == state]

d = c.sort_values('Year')
d
```

```python
plt.figure(figsize = (15, 10))
plt.grid(False)
plt.plot(d['Year'], d['Count'], color = 'r')
plt.xlabel('Year')
plt.ylabel('Occurrences')
plt.title('Occurrences of Anna in CA (1910 - 2014)')
plt.legend(['Occurrences'])
plt.show()
```

```python
plt.figure(figsize = (15, 10))
plt.grid(True)
plt.plot(d['Year'], d['Count'], color = '#e5844b', linestyle = '--', linewidth = 5)        #Search 'color picker' in Google and copy code to apply a desired color
plt.xlabel('Year')
plt.title('Occurrences of Anna in CA (1910 - 2014)')
plt.ylabel('Occurrences')
plt.legend(['Occurrences'])
plt.autoscale(True, 'Year', 'Count')
plt.show()
```

```python
plt.figure(figsize = (15, 10))
plt.grid(True)
plt.plot(d['Year'], d['Count'], color = '#e5844b', linestyle = ':', marker = 'o', label = 'Occurrences', linewidth = 5)
plt.xlabel('Year')
plt.title('Occurrences of Anna in CA (1910 - 2014)')
plt.ylabel('Occurrences')
plt.legend()                                    #Use 'plt.legend(['Occurrences'])' as alternative to 'label = Occurrences' above
plt.show()
```
