# IEX API

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
pd.core.common.is_list_like = pd.api.types.is_list_like
import pandas_datareader
from pandas_datareader import data as wb
import seaborn as sns
sns.set_style('whitegrid')
```

```python
#IEX doesn't provide data more than 5 years prior to code execution date
AAPL = wb.DataReader('AAPL', data_source = 'iex', start = '2019-01-01')['close']
```

```python
AAPL.head()
```

```python
AAPL.tail()
```

```python
AAPL.head(2)
```

```python
AAPL.plot(legend = True, figsize = (25, 15))
plt.title('AAPL 5Y Stock Performance', fontsize = 20, weight = 'bold')
plt.xlabel('Date', fontsize = 15, weight = 'bold')
plt.ylabel('Stock Price ($)', fontsize = 15, weight = 'bold')
plt.legend(prop = {'size': 20})
plt.margins(x = 0)
plt.show()
```

```python
INTC = wb.DataReader('INTC', data_source = 'iex', start = '2019-04-01')['close']
```

```python
INTC.tail()
```
