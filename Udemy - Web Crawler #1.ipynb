```python
from bs4 import BeautifulSoup

html = """
    <html>
       <body>
          <p class='something and nothing but anything'> I'm a paragraph!</p>      #Class must be within opening 'tag'
          <p> Another paragraph!</p>
        </body>
    </html>
"""

doc = BeautifulSoup(html, 'html.parser')                  #Parser cuts through file - 'doc' is an object of BeautifulSoup
```

```python
for p in doc.find_all('p'):                               #Indicating to find all paragraphs
    print(p.text)
    print(p.attrs)
```
