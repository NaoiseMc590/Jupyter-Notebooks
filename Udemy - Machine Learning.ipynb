```python
import pandas as pd
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
file = pd.read_csv('names.csv')
file
```

```python
name = 'Mary'
gender = 'F'
state = 'CA'

df1 = file[file['Name'] == name]
df2 = df1[df1['Gender'] == gender]
df3 = df2[df2['State'] == state]

df4 = df3.sort_values('Year')

x = df4['Year']
y = df4['Count']
```

```python
plt.plot(x, y, color = 'g', linestyle = ':')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression as LinReg
```

```python
x2 = []
print(y)
for value in x:
    x2.append([value])

data = LinReg()

data.fit(x2, y)                                 #'fit' method expects a list of lists (2D format -> [[1910], [1911], [1912]...])
```

```python
data.predict(2020)
```

```python
data.predict([[1920], [1950]])
```

```python
print(x2)
```

```python
preds = data.predict(x2)
```

```python
plt.figure(figsize = (15, 10))
plt.plot(x, y, linewidth = 8)
plt.plot(x, preds, color = 'r', linestyle = '--', linewidth = 4)
plt.title('Occurrences of Mary in CA (1910 - 2014)')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend(['Occurrences', 'Line of Best Fit'])
plt.show()
print(preds)
```
