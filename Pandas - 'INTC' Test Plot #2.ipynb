```python
ex2 = ex.drop(ex.ix[:,'Volume':'Adj_Volume'].head(0).columns, axis= 1)
ex2
```

```python
Intel = ex2.drop(ex2.loc[:,'Open': 'Low'].head(0).columns, axis = 1)
Intel
```

```python
Intel
```

```python
INTC = pd.DataFrame(Intel)
INTC
```

```python
#X-Axis is not optimised -> Problem
plt.figure(figsize = (15, 10))
plt.grid(True)
plt.plot(INTC['Close'], label = 'INTC')
plt.title('Average Stock Price of INTEL CORP. Since IPO')
plt.xlabel('Year')
plt.ylabel('Stock Price')
plt.show()
```

```python
#Convert to DateTime object
INTC2 = pd.DatetimeIndex(INTC['Date'])
INTC2
```

```python
#Group by years -> Solution to lack of X-Axis optimisation
InTel = INTC.groupby([INTC2.year]).mean()
InTel
```

```python
from sklearn.linear_model import LinearRegression as LinReg
```

```python
x = InTel.index.values.reshape(-1, 1)
y = InTel['Close'].values
```

```python
reg = LinReg()
reg.fit(x, y)
prediction = reg.predict(x)
print('Accuracy: ' + str(reg.score(x, y)))                          #Not very accurate prediction          
```

```python
plt.figure(figsize = (25, 15))
plt.grid(True)
plt.title('Linear Regression - INTC')
plt.xlabel('Year')
plt.ylabel('Stock Price')
plt.scatter(x= x, y= prediction, color = 'b')
plt.scatter(x= x, y= y, c= 'black')
#plt.plot()
plt.show()
```
