```python
x = data[['House Size (sq.ft.)', 'Number of Rooms', 'Year of Construction']] #Double brackets to make two-dimensional
y = data['House Price']
```

```python
x1 = sm.add_constant(x)
reg = sm.OLS(y, x1).fit()
reg.summary()
```
