```python
#Facial Recognition
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
#EXAMPLE 1
img = cv2.imread('img.jpg')
img
```

```python
i = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(i)
plt.show()
```

```python
a = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(a, 'gray')
plt.show()
```

```python
#Classifier -> Set of data which knows what a face looks like etc. ('haarcascades' folder)
classifier = cv2.CascadeClassifier('../data/haarcascades/haarcascade_frontalface_alt2.xml')
faces = classifier.detectMultiScale(a)      #'detesctMultiScale' covers dfferent areas to ensure accuracy
print(faces)                                #Prints positions of the faces identified
```

```python
b = img.copy()
for face in faces:
    x, y, w, h = face
    cv2.rectangle(b, (x,y), (x+w, y+h), (0, 255, 0), 50)
    print(face)
    
i = cv2.cvtColor(b, cv2.COLOR_BGR2RGB)
plt.imshow(i)
plt.show()                                                                       #Notice inaccuracies
```

```python
classifier = cv2.CascadeClassifier('../data/haarcascades/haarcascade_frontalface_alt2.xml')
faces = classifier.detectMultiScale(a, minNeighbors = 1)

b = img.copy()
for face in faces:
    x, y, w, h = face
    cv2.rectangle(b, (x,y), (x+w, y+h), (0, 255, 0), 50)
    print(face)
    
i = cv2.cvtColor(b, cv2.COLOR_BGR2RGB)
plt.imshow(i)
plt.show()                                                      #Notice greater quantity of inaccuracies            
```

```python
classifier = cv2.CascadeClassifier('../data/haarcascades/haarcascade_frontalface_alt2.xml')
faces = classifier.detectMultiScale(a, minNeighbors = 10)

b = img.copy()
for face in faces:
    x, y, w, h = face
    cv2.rectangle(b, (x,y), (x+w, y+h), (0, 255, 0), 50)
    print(face)
    
i = cv2.cvtColor(b, cv2.COLOR_BGR2RGB)
plt.imshow(i)
plt.show()                                            #Only one face recognised due to 'minNeighbors = 10' above
```

```python
#EXAMPLE 2
image = cv2.imread('id1_edited-1.jpg')
image
```

```python
e = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(e)
plt.show()
```

```python
l = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(l, 'gray')
plt.show()
```

```python
classifier = cv2.CascadeClassifier('../data/haarcascades/haarcascade_frontalface_alt2.xml')
faces2 = classifier.detectMultiScale(e, minNeighbors = 1)

p = image.copy()
for face in faces2:
    x, y, w, h = face
    cv2.rectangle(p, (x,y), (x+w, y+h), (0, 0, 0), 6)                        #Dimensions/Instructions for drawing of rectangle
    print(face)
    
e = cv2.cvtColor(p, cv2.COLOR_BGR2RGB)
plt.imshow(e)
plt.show()                  
```
